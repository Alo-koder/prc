```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
FILENAME = 'T:\\Team\\Szewczyk\\Data\\2024-03-05\\A02506.dat'
```

```python
data = pd.read_csv(FILENAME, skiprows=5, sep='\t')
```

```python
data
```

```python
data = data.rename(columns = {
    'Time [s]'  : 't',
    'U [V]'     : 'U',
    'I [A]'     : 'I',
    'EMSI [%]'  : 'emsi',
})
data = data.drop(columns = 'Ill [V]')
data
```

```python
fit = np.polyfit(data.t, data.emsi, 1)
emsi_corrected = data.emsi - np.polyval(fit, data.t)
plt.plot(data.t, emsi_corrected)
data.emsi = emsi_corrected
data.I = -data.I
```

```python
%matplotlib qt
```

```python
fig, axs = plt.subplots(2, sharex = True)
axs[0].plot(data.t, data.I)
axs[0].set_title('current [A]')
axs[1].plot(data.t, data.emsi)
axs[1].set_title('emsi [%]')
```
