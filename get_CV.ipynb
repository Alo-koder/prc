```python
import eclabfiles as ecf
import numpy as np
from matplotlib import pyplot as plt
```

```python
date='24 X'
```

```python
filenames   = {
    '24 X': 'T:\\Team\\Szewczyk\\Data\\20231024\\raw\\A00100_C01.mpr',
    }
```

```python
def read_data(filename=filenames[date]):
    
    raw_data = ecf.to_df(filename)
    
    # Renaming columns
    col_dict = {'time': 't', 'Ewe': 'V', '<I>': 'I'}
    data = raw_data.rename(columns = col_dict)

    return data

data = read_data()
```

```python
def get_CV(data=data):

    voltage = np.array(data['V'])
    current = np.array(data['I'])
    # time = np.array(data['t'])

    fig, ax = plt.subplots()
    plt.title(f'CV {date}')
    plt.xlabel('Voltage')
    plt.ylabel('I [check the unit]')
    # plt.scatter(voltage, current, c=time, s=0.2)
    plt.plot(voltage, current)

get_CV(data)
```
